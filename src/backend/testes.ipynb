```python
from gerador import *
import pandas as pd
import numpy as np
from time import sleep, time
import matplotlib.pyplot as plt
from random import choice, randint, uniform
from server import NovoData, Qnts
```

```python
def gerarTeste(pvrs:int=5,clss:int=10,srvs:int=10):
    return NovoData(
        name=f'teste_{pvrs}_{clss}_{srvs}',
        qnts=Qnts(prvs=pvrs, clss=clss, srvs=srvs),
        classes=[
            {
                'categoria': randint(0,9),
                'cpus': [ randint(0,11) for _ in range(randint(1,3)) ],
                'ram': [ randint(0,9) for _ in range(randint(1,3)) ],
                'hd': {
                    'min': randint(0,50),
                    'max': randint(50,100)
                },
                'disponibilidade': uniform(1, 99.9999),
                "tempoResposta": 1.0,
                "custo": 679.14
            } for _ in range(clss)
        ]
    )

    # return {
    #     'name': f'teste_{pvrs}_{clss}_{srvs}',
    #     'qnts': {
    #         'prvs': pvrs,
    #         'clss': clss,
    #         'srvs': srvs
    #     },
    #     'classes': [
    #         {
    #             'categoria': randint(0,9),
    #             'cpus': [ randint(0,11) for _ in range(randint(1,3)) ],
    #             'ram': [ randint(0,9) for _ in range(randint(1,3)) ],
    #             'hd': {
    #                 'min': randint(0,50),
    #                 'max': randint(50,100)
    #             },
    #             'disponibilidade': uniform(1, 99.9999),
    #             "tempoResposta": 1.0,
    #             "custo": 679.14
    #         } for _ in range(clss)
    #     ]
    # }
```

```python
resultados = []
padroes = [1, 5, 10, 25, 50]
 
```

```python
def run():
    for prov in padroes:
        for clss in range(1, 11):
            for serv in padroes:
                inicio = time()
                print(f'Gerando teste {prov} {clss} {serv}')
                _ = sequencia(gerarTeste(prov, clss, serv))
                fim = time()
                resultados.append(
                    {
                        'qntPrv': prov,
                        'qntClss': clss,
                        'qntSrvs': serv,
                        'tempo': fim - inicio
                    }
                )
    # for t in testes:
    #     inicio = time()
    #     print(f'Gerando teste {t}')
    #     _ = sequencia(gerarTeste(*t))
    #     fim = time()
    #     resultados.append(
    #         {
    #             'qntPrv': t[0],
    #             'qntClss': t[1],
    #             'qntSrvs': t[2],
    #             'tempo': fim - inicio
    #         }
    #     )
```
